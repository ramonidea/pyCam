```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
### resultfps(1), resultlat(1) -> ROS Raw Image and Depth. (1) -> Depth
### comfps(1), comlat(1) -> ROS Compressed Image and Depth (1) -> Depth
```

```python
with open("resultfps.json","r") as inputfile:
    fps = json.load(inputfile)
with open("resultLat.json","r") as inp:
    lat = json.load(inp)
with open("resultfps1.json","r") as inputfile:
    fps1 = json.load(inputfile)
with open("resultLat1.json","r") as inp:
    lat1 = json.load(inp)
    
    
```

```python
def lineplot(x,y,x_label="",y_label="",title=""):
    _, ax = plt.subplots()
    ax.plot(x,y,lw=2,color="#539caf",alpha=1)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
```

```python
lineplot(np.arange(0,505,5),fps, "Sec", "FPS","Color FPS Rate")
lineplot(np.arange(0,505,5),fps1, "Sec", "FPS","Depth FPS Rate")
lineplot(np.arange(0,505,5),np.multiply(lat,1000), "ms", "Latency(s)","Color Latency")
lineplot(np.arange(0,505,5),np.multiply(lat1,1000), "ms", "Latency(s)","Depth Latency")
```
